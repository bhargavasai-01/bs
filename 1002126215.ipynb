### 5. Spot outliers and what to do with them?

```python
def graph(y):
    sns.boxplot(x="Species", y=y, data=df_fish)
 
plt.figure(figsize=(10,10))
     
# Adding the subplot at the specified
# grid position
plt.subplot(221)
graph('Length1')
 
plt.subplot(222)
graph('Width')
 
plt.subplot(223)
graph('Length2')
 
plt.subplot(224)
graph('Height')
 
plt.show()
```

this is box plot which gives outliers and range of data

```python
sns.boxplot(x='Width', data=df_fish)
```

the width feature have no outlier 

```python
sns.boxplot(x='Height', data=df_fish)
```

the height have no outliers

```python
sns.boxplot(x='Length1', data=df_fish)
```

the length1 have 2 outliers

```python
sns.boxplot(x='Length2', data=df_fish)
```

the length2 have 2 outliers

```python
sns.boxplot(x='Length3', data=df_fish)
```

the length3 have 1 outliers

```python
Q1 = np.percentile(df_fish['Length3'], 25,
                interpolation = 'midpoint')
 
Q3 = np.percentile(df_fish['Length3'], 75,
                interpolation = 'midpoint')
IQR = Q3 - Q1
 
print("Old Shape: ", df_fish.shape)
 
# Upper bound
upper = np.where(df_fish['Length3'] >= (Q3+1.5*IQR))
 
# Lower bound
lower = np.where(df_fish['Length3'] <= (Q1-1.5*IQR))
 
# Removing the Outliers
df_fish.drop(upper[0], inplace = True)
df_fish.drop(lower[0], inplace = True)
 
print("New Shape: ", df_fish.shape)
 
sns.boxplot(x='Length3', data=df_fish)
```
